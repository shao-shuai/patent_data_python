```python
import numpy as np
import pandas as pd
import requests
import json
```

```python
#将要访问的URL赋给一个变量，这个URL是用来获取4月14日发生的PTAB事件，如IPR，PGR等
#关于这个API的具体文档请访问这里 https://ptabdata.uspto.gov/ptab-api/swagger
url = "https://ptabdata.uspto.gov/ptab-api/trials?filingDate=2019-04-11"
```

```python
#使用get请求数据，然后将其转换成jason数据格式
ptab = requests.get(url).json()
ptab
```

```python
#上面json的格式太乱，我们调整一下格式输出
content = json.dumps(ptab, indent = 4, sort_keys=True)
print(content)
```

```python
#将json格式转换成pandas中的DataFrame格式
pd.DataFrame(ptab['results'])
```
