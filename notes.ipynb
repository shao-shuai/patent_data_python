```python
#construct API URL
#url = "https://ptabdata.uspto.gov/ptab-api/trials?filingDate=2019-04-24&offset=0"
url = "https://ptabdata.uspto.gov/ptab-api/trials?filingDateFrom=2019-03-01&filingDateTo=2019-04-28"
```

```python
#request data
ptab = requests.get(url).json()
ptab
```

```python
#generate dataframe
ptab_df = pd.DataFrame(ptab['results'])
ptab_nolin= ptab_df.loc[:, ptab_df.columns!='links']
#ptab_nolin = ptab_nolin.sort_values(by='filingDate')
ptab_nolin
```

```python
#link patent number to google patent
def make_clickable(number):
    return '<a href=https://patents.google.com/patent/US{}>{}</a>'.format(number, number)
df = ptab_nolin.drop(['petitionerPartyName', 'trialName'], axis = 1)
df.style.format({'patentNumber':make_clickable})
```
