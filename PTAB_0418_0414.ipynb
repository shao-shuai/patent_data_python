```python
#create API URL
url = "https://ptabdata.uspto.gov/ptab-api/trials?filingDateFrom=2019-04-08&filingDateTo=2019-04-14&limit=100"
```

```python
#使用get请求数据，然后将其转换成jason数据格式
ptab = requests.get(url).json()
```

```python
ptab_df = pd.DataFrame(ptab['results'])
ptab_nolin= ptab_df.loc[:, ptab_df.columns!='links']
```

```python
def make_clickable(number):
    return '<a href=https://patents.google.com/patent/US{}>{}</a>'.format(number, number)

ptab_nolin.style.format({'patentNumber':make_clickable})
```
